```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Input,InputLayer,Dropout
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.neighbors import LocalOutlierFactor
```

```python

ACCURACY_THRESHOLD = 0.99

class myCallback(tf.keras.callbacks.Callback):
	def on_epoch_end(self, epoch, logs={}):
		if(logs.get('accuracy') > ACCURACY_THRESHOLD):
			print("\nReached %2.2f%% accuracy, so stopping training!!" %(ACCURACY_THRESHOLD*100))
			self.model.stop_training = True
callbacks = myCallback()
```

```python
final_data=pd.read_csv(r'final_data.csv',index_col=0)
```

```python
final_data
```

```python
X_train, X_test, y_train, y_test=train_test_split(final_data.iloc[:,:80],final_data['label'],test_size=0.25,shuffle=True)
```

```python
y_train=to_categorical(y_train,5)
y_test =to_categorical(y_test, 5)
```

```python
X_train=np.array(X_train)
```

```python
model=Sequential([
                  InputLayer(80),
                  Dense(256,activation='relu'),
                  Dense(128,activation='tanh'),
                  Dense(5,activation='softmax'),

])
model.compile(optimizer='adam',metrics=['accuracy'],loss='categorical_crossentropy',)
model.fit(X_train,y_train,epochs=20,callbacks=[callbacks],steps_per_epoch=10)
```

```python
model.evaluate(X_test,y_test,verbose=2) #97%
```
